```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('preprocessed_diamond.csv')
```

Univariate analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

fig, axes = plt.subplots(1, 1, figsize=(8, 5))

fig.suptitle('UNIVARIATE ANALYSIS - Distribution of Variables', fontsize=16, y=1.02)

axes.hist(df['discount'], bins=50, color='skyblue', edgecolor='black')

axes.axvline(df['discount'].mean(), color='red', linestyle='--', label='Mean')
axes.axvline(df['discount'].median(), color='green', linestyle='--', label='Median')

axes.set_title('Discount Distribution')
axes.set_xlabel('Discount (%)')
axes.set_ylabel('Frequency')
axes.legend()

plt.tight_layout()
plt.show()
```

Multi variate Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group by color and clarity, calculate average discount
grouped = df.groupby(['color', 'clarity'])['discount'].mean().unstack()

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(grouped, annot=True, fmt=".2f", cmap="YlGnBu", cbar_kws={'label': 'Avg Discount (%)'})

plt.title('Average Discount by Color and Clarity')
plt.xlabel('Clarity')
plt.ylabel('Color')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group by quality_score and category_count, compute average discount
grouped = df.groupby(['quality_score', 'category_count'])['discount'].mean().unstack()

# Plot heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(grouped, annot=False, fmt=".2f", cmap="YlOrBr", cbar_kws={'label': 'Avg Discount (%)'})

plt.title('Average Discount by Quality Score and Category Count')
plt.xlabel('Category Count')
plt.ylabel('Quality Score')
plt.tight_layout()
plt.show()
```
